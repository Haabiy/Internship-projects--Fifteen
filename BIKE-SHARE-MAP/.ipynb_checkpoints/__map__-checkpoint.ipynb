### <font color='Orange'>Reading and cleaning data</font> 

```python
def get_bike_share_map():
    
    headers = {
    'Accept': '*/*',
    'Accept-Language': 'fr-FR,fr;q=0.9,en-US;q=0.8,en;q=0.7',
    'Connection': 'keep-alive',
    'Origin': 'https://bikesharemap.com',
    'Referer': 'https://bikesharemap.com/',
    'Sec-Fetch-Dest': 'empty',
    'Sec-Fetch-Mode': 'cors',
    'Sec-Fetch-Site': 'cross-site',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36',
    'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="101", "Google Chrome";v="101"',
    'sec-ch-ua-mobile': '?0',
    'sec-ch-ua-platform': '"Windows"',
    }

    params = {'unused': 'colour=0',}

    response = requests.get('https://julie.geog.ucl.ac.uk/~ollie/bikesapi/systems_js.php', params=params, headers=headers)
    
    return response.json()
```

```python
# Create a DataFrame
bike_map = get_bike_share_map()
df = pd.DataFrame(bike_map)

# Transpose columns into rows
df_t = df.T
df_t.head(10)
#df.head()

# Save as a .csv file
#df.to_csv('final.csv')
```

```python
# Let's use country names as indices, that way, it's easier to target countries and manipulate their data
df_t = df_t.reset_index(drop=True)
df_t.set_index('country_name', inplace=True)
dff = pd.DataFrame(df_t)
dff.head()
```

```python
# Search for a city; NB : city_name is a column
def search_city(city):
    get_city = dff[dff['city_name']==city]
    #Save the data as a .csv file
    #search.to_csv("Barcelona.csv")
    return get_city
```

```python
# Unlike city_name, country_name is an index which could be accessed through DataFrame.index
def search_country(country):
    get_country = dff[(dff.index == country)]
    return get_country
```

```python
# In case you're not sure about the spelling of the country name
def search_str_country(string):
    get_string = dff[(dff.index).str.contains(string)]
    return get_string
```

```python
# The same is true for city names
# In case you're not sure about the spelling of the city name
def search_str_city(string):
    get_string = dff[(dff.city_name).str.contains(string)]
    return get_string
```

```python
# Let's narrow down the data by only focusing on the info I'm interested in
# I'm interested in getting the city name, number of bikes available, name of the city hall, launch date, closed date, operator name, system type, system website and the technology they use 
def df_narrowed(dtframe):
    _df = dtframe[['city_name', 'nominal_num_bikes', 'cityhall', 'launch_date', 'closed_date', 'timezone', 'operator', 'operator_name', 'system_type', 'system_website', 'technology']]
    return _df

def get_df_narrowed(country):
    df_ = pd.DataFrame(df_narrowed(dff))
    x = df_[df_.index == country]
    return x
```

# <font color='Orange'>Data analysis and visual representation</font> 

```python
# Next, let's have a look at country's mobility technology based on the nominal number of bikes they have
def bike_tech(country):
    y = get_df_narrowed(country).sort_values(by = ['nominal_num_bikes'], ascending = False)
    return y
```

```python
# see, the dominant technology in Germany is nextbike
bike_tech('Germany')
```

```python
# Top ten countries in descending order of nominal number of bikes
# Most of the top twenty technologies are in the cities of China

def top_tech_plot(x,y,i):
    
    plt.rcParams["figure.figsize"] = (8,8)
    plt.rcParams["figure.autolayout"] = True
    fig = plt.figure()
    ax = fig.add_subplot(2,1,i)
    barchart = ax.barh(x, y)
    ax.set_xlabel('Nominal number of bikes', labelpad=7)
    if(i==1):
        ax.set_ylabel('Name of cities')
    else:
        ax.set_ylabel('Name of operators')
    ax.yaxis.label.set_color('blue')
    ax.xaxis.label.set_color('blue')
    
    return barchart     
```

```python
top_df = df_narrowed(dff).sort_values(by = ['nominal_num_bikes'], ascending = False).head(10)
x = top_df.city_name.head(10)
y = top_df.nominal_num_bikes.head(10)
z = top_df.operator_name.head(10)
top_df
```

```python
# plot 1 : x = name of cities
# plot 2 : x = name of operators
# y remains being the nominal number of bikes
top_tech_plot(x,y,1)
top_tech_plot(z,y,2)
plt.show()
```

```python
non_empty_file = np.array([])
def save_df(sister_city, lst_len):
    # Instead of retrieving all the information, let's narrow down by the function made earlier
    _return = df_narrowed(dff)[(df_narrowed(dff).city_name).str.contains(sister_city)]
    i = 0
    # file_ : is the general file I used as a reference to filter out the targeted cities
    # first save each as .csv file
    while(i<=lst_len):
        file_ = _return.to_csv(sister_city+".csv")
        i = i + 1
    # filename is in the same directory
    filename = "./"+sister_city+".csv"
    with open(filename, encoding='utf-8') as f:
        readl = f.readlines()
        try:
            # we are only interested in cities that are present on the bike share map
            # If the length is > 1, that means they contain information as they're present on the map, otherwise; pass
            # if the length is 1, they only contain the name of columns (city_names, nominal_num_bikes...)
            if(len(readl)>1):
                global file 
                file = _return.to_csv("__"+sister_city+".csv")
                np.append(non_empty_file, sister_city)
                print("__"+sister_city+".csv", ",", len(readl)-1, "x 9")
            else:
                pass
        except Exception:
            pass
    return file
```

```python
lst_len = len(lst_city)
for cities in range(lst_len):
    save_df(lst_city[cities], lst_len)
```

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
url  = 'https://www.aznations.com/population/ua/cities/luhansk'
driver.get(url)
pop_num = []
class ff():
    def get_search_result(self, city):
        search = driver.find_element(By.XPATH, '/html/body/header/div/div/div[3]/div/form/div/span/input')
        self.search = search
        self.search.send_keys(Keys.CONTROL + 'a', Keys.BACKSPACE)
        # Checking if the text box has been cleared
        #print("__________________XXXX________________")
        self.city = city
        self.search.send_keys(self.city)
        self.search.send_keys(Keys.ENTER)
        #driver.maximize_window()
        sleep(2)

    def get_num(self):
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        try:
            for body in soup.find_all('div', {'class' : 'page-heading-title'}):
                span = body.find_all('span')[0].text.strip()
                num = span[:-7]
                pop_num.append(self.city + '| ' +num)
                print(num)
        except Exception:
            x = input('Has '+self.city+' been registered on the website?')
            if(x=='yes'):
                self.get_num()
            else:
                pop_num.append(self.city+'| '+'None')
                sleep(1)
                pass
```

```python
get = ff()
for i in lst_city:
    get.get_search_result(i)
    get.get_num()
    sleep(1)
```
